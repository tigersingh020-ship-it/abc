## Python Loops

Often we need to perform the same operation many times.

Instead of writing the same code repeatedly, we use **loops**.

A loop lets us run the same block of code for every item in a list.

Structure of a loop:

for item in collection:
    do something

````python
prices = [100, 102, 101, 105]

for price in prices:
    print(price)
````

````python
for x in prices:
    print(x)
````

## 10. Data Cleaning

Real-world data is often messy, so we clean it before analysis.

Common steps:
- keep only the column we need
- rename the column clearly
- remove missing values

We usually use the **Close** price.

````python
data = data[["Close"]]
data = data.rename(columns={"Close": "price"})
data = data.dropna()

data.head()
````

## 11. Compute Daily Returns

A **return** tells us the percentage change in price from one day to the next.

Formula:

\[
return = \frac{price_t - price_{t-1}}{price_{t-1}}
\]

Pandas gives us a built-in function:

`pct_change()`

````python
data["return"] = data["price"].pct_change()
data.head()
````

## 12. Compute Cumulative Returns

Cumulative return shows the growth of $1 invested over time.

We compute it using:

`(1 + return).cumprod()`

````python
data["cum_return"] = (1 + data["return"]).cumprod()
data.tail()
````

## 13. Rolling Volatility

Volatility measures how much returns move around.

A common way to estimate volatility is:

`standard deviation of returns`



100 → 101 → 102 → 101 → 102

100 → 120 → 90 → 130 → 80

Volatility = how much prices move up and down

````python
import matplotlib.pyplot as plt

stable = [100,101,102,101,102]
volatile = [100,120,90,130,80]

plt.plot(stable, label="Low Volatility")
plt.plot(volatile, label="High Volatility")

plt.legend()
plt.title("Volatility Comparison")
plt.show()
````

A **rolling volatility** means we compute the volatility over the last N days.

For example:
- 20 trading days is about 1 month

````python
data["rolling_vol_20"] = data["return"].rolling(window=20).std()
data.tail()
````

## Summary Statistics Table

After computing returns and volatility, we often want a quick summary of the data.

Instead of looking at thousands of rows, we compute key statistics such as:

- mean return
- volatility (standard deviation of returns)
- maximum return
- minimum return

This gives us a quick overview of the stock's behavior.

### Key Statistics

We will compute:

Mean Return → average daily return  
Volatility → standard deviation of returns  
Max Return → largest daily return  
Min Return → smallest daily return

## Creating a Summary Dictionary

Instead of computing these one by one, we can store them in a dictionary.

This makes our code cleaner.

````python
summary = {
    "mean_return": data["return"].mean(),
    "volatility": data["return"].std(),
    "max_return": data["return"].max(),
    "min_return": data["return"].min()
}

summary
````

## Convert Dictionary to a Table

We can convert the dictionary into a Pandas DataFrame.

````python
summary_df = pd.DataFrame([summary])

summary_df
````
