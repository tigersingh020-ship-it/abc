- A plan with measurements, budget, and style preferences is specific.
- Code is the computer’s version of a clear plan.

## 4) Variables (labeled boxes)

- Variables store information so we can reuse it.
- Think of variables as labeled boxes (budget box, size box, style box).

Run the cell and then change the values (e.g., budget) and re-run.


```python
budget = 5000
room_size_sqft = 200
style = "modern"

print("Budget:", budget)
print("Room size (sqft):", room_size_sqft)
print("Style:", style)
```

## 5) User input (like a website form)

- Websites ask users questions (forms).
- `input()` lets us collect user answers.
- `int(...)` converts text to a number when needed.

Run and enter your own values.


```python
name = input("What is your name? ")
budget = int(input("Enter your budget (number): "))
style = input("Enter your style (e.g., modern/minimal/classic): ")

print("\nHi", name)
print("Budget:", budget)
print("Style:", style)
```

## 6) Data structures (organizing information)

- A **list** stores multiple related items (like a mood board list).
- A **dictionary** stores labeled information (like a design spec sheet).

Run the cells and discuss why these are useful for websites and AI.


```python
styles = ["modern", "minimal", "classic"]
print("Available styles:", styles)
```

```python
room = {
    "name": name,
    "budget": budget,
    "style": style,
    "size_sqft": 200
}
print("Room spec sheet:", room)
```

## 7) If / Else logic (website decisions)

- Websites constantly make decisions:
  - show this message or that message
  - allow something or block it
- `if/else` is how we encode decisions.

Example: give a simple recommendation based on budget.


```python
if budget < 1000:
    print("Recommendation: Focus on paint + small decor upgrades.")
elif budget < 5000:
    print("Recommendation: Add lighting + a key furniture piece.")
else:
    print("Recommendation: You can consider a fuller room refresh!")
```

## 8) Loops (repeat tasks efficiently)

- Loops help us repeat actions without rewriting code.
- A website might loop through posts, products, or recommendations.

Example: loop through a list of furniture and print each item.


```python
furniture = ["sofa", "coffee table", "lamp", "rug"]
for item in furniture:
    print("Suggested item:", item)
```


- AI models don’t “see rooms” — they see **organized data** (numbers + categories).
- Today we built the exact kinds of structures that AI and websites use.
- Later, instead of rules like `if/else`, we’ll train a model from examples.

Organized data → model → prediction → show result on website.
